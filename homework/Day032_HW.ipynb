### 像 test/test2.png 這種複雜的圖片怎麼辦？

PIL九種模式:
* '1': 二值圖像, 非黑即白
* 'L': 灰色圖像
* 'P': 8位彩色圖像
* 'RGB': 
* 'RGBA': 32位彩色圖像
* 'CMYK': 32位彩色圖像
* 'YCbCr': 24位彩色圖像
* 'I': 32位整型灰色圖像
* 'F': 32位浮點灰色圖像


```python
img = Image.open("test/test2.png").convert("L")
img
```

```python
code = pytesseract.image_to_string(img)
print(code)
```
