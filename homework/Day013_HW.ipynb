### ① 印出文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

```python
import pandas as pd

info_all = []
for d in soup.find_all(class_="title"):
    try:
        url = 'https://www.ptt.cc/' + d.find('a')['href']
        # print(url)
        soup_2 = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
        info = []
        for dd in soup_2.find_all(class_='article-meta-value'):
            info.append(dd.text)
        info_all.append(info)
    except:
        continue
df = pd.DataFrame(info_all, columns=['作者', '看板', '標題', '時間'])
df
```

### ③ 試著爬爬看其他版的文章

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/stock/index.html'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

```python
import pandas as pd

info_all = []
for d in soup.find_all(class_="title"):
    try:
        url = 'https://www.ptt.cc/' + d.find('a')['href']
        # print(url)
        soup_2 = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
        info = []
        for dd in soup_2.find_all(class_='article-meta-value'):
            info.append(dd.text)
        info.append(url)
        info_all.append(info)
    except:
        continue
df = pd.DataFrame(info_all, columns=['作者', '看板', '標題', '時間', 'URL'])
df
```
