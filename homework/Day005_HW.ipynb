```python
import requests
```

```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
p_res = r.text
p_res
```

```python
import json

p_data = json.loads(p_res)
print("è³‡æ–™æ•¸é‡: {}\næ¬„ä½: {}".format(len(p_data), [i for i in p_data[0]]))
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
p_total_commentCount = 0
p_total_likeCount = 0
for n, i in enumerate(p_data):
    print("{:2d}. [æ¨™é¡Œ]:{} [è²¼æ–‡æ™‚é–“]:{} [ç•™è¨€äººæ•¸]:{} [æŒ‰è®šäººæ•¸]:{}".format(n+1, i['title'], i['createdAt'], i['commentCount'], i['likeCount']))
    p_total_commentCount += int(i['commentCount'])
    p_total_likeCount += int(i['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
np_res = r.text
np_data = json.loads(np_res)
print("è³‡æ–™æ•¸é‡: {}\næ¬„ä½: {}".format(len(np_data), [i for i in np_data[0]]))
```
