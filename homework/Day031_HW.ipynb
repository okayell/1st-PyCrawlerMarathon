### 尋找一個網站是有作 Header 檢查的（Hint: 有加跟沒有加回傳結果不一樣）

```python
url = 'https://www.coursera.org/learn/ntumlone-mathematicalfoundations'
```

```python
import requests
r = requests.get(url)
response = r.text

print(len(response))
response
```
